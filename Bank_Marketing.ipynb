```python
marital_values = list(data['marital'].value_counts().index)
marital = data['marital'].value_counts()
marital = 100*marital/marital.sum()
plt.bar(marital_values,marital.values)
plt.show()
```

```python
education_values = list(data['education'].value_counts().index)
education = data['education'].value_counts()
education = 100*education/education.sum()
plt.bar(education_values,education.values)
plt.xticks(rotation=90)
plt.show()
```
